# Profitability


ROA

```python
for i in range(11,20):
    annual_data['ROA'+str(i)] = (annual_data['Total income from continuing operations'+str(i)]-annual_data['Other income & extra-ordinary income'+str(i)])/annual_data['Total assets'+str(i-1)]
```

CFO


```python
for i in range(11,20):
    annual_data['CFO'+str(i)] = (annual_data['Net cash flow from operating activities'+str(i)])/annual_data['Total assets'+str(i-1)]
```

```python
ROA
```

Delta ROA

```python
ROA = annual_data['ROA11']
for i in range(12,20):
    ROA = pd.concat([ROA,annual_data['ROA'+str(i)]],axis=1) 
```

```python

del_ROA=ROA.diff(axis=1).drop('ROA11',axis=1)
```

Accruals

```python
CFO = annual_data['CFO11']
for i in range(12,20):
    CFO = pd.concat([CFO,annual_data['CFO'+str(i)]],axis=1) 
```

```python
accrual=[]
accrual=pd.DataFrame(accrual)
for i in range(11,20):
    accrual['accrual'+str(i)] = ROA['ROA'+str(i)] - CFO['CFO'+str(i)]
```

# Cap Str

Delta Leverage

```python
for i in range(11,20):
    annual_data['Leverage'+str(i)] = (annual_data['Long term borrowings excl current portion'+str(i)])/annual_data['Total assets'+str(i-1)]
```

```python
Leverage = annual_data['Leverage11']
for i in range(12,20):
    Leverage = pd.concat([Leverage,annual_data['Leverage'+str(i)]],axis=1) 
```

Delta Leverage

```python
del_Leverage=Leverage.diff(axis=1).drop('Leverage11',axis=1)
```

```python
del_Leverage
```

Current Ratio

```python
for i in range(11,20):
    annual_data['CR'+str(i)] = (annual_data['Current assets'+str(i)])/annual_data['Current liabilities'+str(i)]
```

```python
Liquid = annual_data['CR11']
for i in range(12,20):
    Liquid = pd.concat([Liquid,annual_data['CR'+str(i)]],axis=1) 
```

```python
del_Liquid=Liquid.diff(axis=1).drop('CR11',axis=1)
```

EQ_Offer

```python
for i in range(11,20):
    temp = annual_data['Issue of fresh capital'+str(i)]==0
    annual_data['EQ_offer'+str(i)] = temp.astype(int)
```

# Operational Efficiency

Margin

```python
for i in range(11,20):
    annual_data['Margin'+str(i)] = annual_data['Net Profit/(Loss) for the period from continuing operations (after tax)'+str(i)]/annual_data['Sales'+str(i)]
```

```python
Margin = annual_data['Margin11']
for i in range(11,20):
    Margin = pd.concat([Margin,annual_data['Margin'+str(i)]],axis=1) 
```

```python
del_Margin = Margin.diff(axis=1).drop('Margin11',axis=1)
```

TurnOver

```python
for i in range(11,20):
    annual_data['Turnover'+str(i)] = annual_data['Sales'+str(i)]/annual_data['Total assets'+str(i-1)]
```

```python
TurnOver = annual_data['Turnover11']
for i in range(11,20):
    TurnOver = pd.concat([TurnOver,annual_data['Turnover'+str(i)]],axis=1) 
```

```python
del_turnover = TurnOver.diff(axis=1).drop('Turnover11',axis=1)
```

```python
del_turnover
```

```python
accrual
```

```python
for i in range(12,20):
    temp = ROA['ROA'+str(i)]>0
    annual_data['ROA_score'+str(i)] = temp.astype(int)
    
    temp = CFO['CFO'+str(i)]>0
    annual_data['CFO_score'+str(i)] = temp.astype(int)
    
    
    temp = del_ROA['ROA'+str(i)]>0
    annual_data['del_ROA_score'+str(i)] = temp.astype(int)
    
    
    temp = del_Leverage['Leverage'+str(i)]<0
    annual_data['Leverage_score'+str(i)] = temp.astype(int)
    
    
    temp = del_Liquid['CR'+str(i)]>0
    annual_data['CR_score'+str(i)] = temp.astype(int)
    
    temp = accrual['accrual'+str(i)]<0
    annual_data['accrual_score'+str(i)] = temp.astype(int)
    
    
    temp = del_Margin['Margin'+str(i)]>0
    annual_data['Margin_score'+str(i)] = temp.astype(int)
    
    
    temp = del_turnover['Turnover'+str(i)]>0
    annual_data['Turnover_score'+str(i)] = temp.astype(int)
    
```

```python
for i in range(12,20):
    annual_data['Fscore'+str(i)] = annual_data['ROA_score'+str(i)]+annual_data['CFO_score'+str(i)]+annual_data['del_ROA_score'+str(i)]+annual_data['Leverage_score'+str(i)]+annual_data['CR_score'+str(i)]+annual_data['accrual_score'+str(i)]+annual_data['Margin_score'+str(i)]+annual_data['Turnover_score'+str(i)]+annual_data['EQ_offer'+str(i)]
```

RETURNS FOR FIRMS WITH FSCORE 8 or ABOVE

```python
ret_long = []
for i in range(12,19):
    z = annual_data[annual_data['Fscore'+str(i)]>7]
    returns=(z['Closing Price13']-z['Closing Price12'])/z['Closing Price12']
    ret_long.append(returns.mean())
```

RETURNS FOR FIRMS WITH FSCORE 2 OR BELOW

```python
ret_short = []
for i in range(12,19):
    z = annual_data[annual_data['Fscore'+str(i)]<3]
    returns=(z['Closing Price13']-z['Closing Price12'])/z['Closing Price12']
    ret_short.append(returns.mean())
```

HEDGE PORTFOLIO RETURN

```python
Years = ['12-13','13-14','14-15','15-16','16-17','17-18','18-19']
Returns = pd.DataFrame(ret_long,index = Years,columns = ['long'])
```

```python
frames = [Returns,pd.DataFrame(ret_short,index=Years,columns=['short'])]
Returns = pd.concat(frames,axis=1)
```

```python
Returns['Hedge'] = Returns['long']-Returns['short']
```
